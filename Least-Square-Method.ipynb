```python
# b = (nΣ(xiyi) – Σ(xi)Σ(yi)) ÷ (nΣ(xi2)-Σ(xi)2) 
# a = ȳ – b.x̄ 

```

```python
X = [95, 85, 80, 70, 60 ]
Y = [90, 80, 70, 65, 60 ]
n = len(X)
```

```python
def calculate(x,y,n):
  sumX = sum(x)
  sumY = sum(y)
  sqX = 0 # Σ(xiyi)
  SsqX = 0 # Σ(xi2)
  for i in range (n):
    sqX+= x[i]*y[i]
    SsqX+=x[i]*x[i]
  b =(n * sqX - sumX*sumY)/ (n*SsqX - sumX*sumX)
  a =(sumY/n) - b*(sumX/n)
  print("Y = ", '%.3f'%a, " + ", '%.3f'%b, "*X", sep="")
  return a,b
[a,b] = calculate(X,Y,n)
print(a)
print(b)
```

```python
def pred(a,b,n,x):
  y_pred = [0]*n
  for i in range (n):
    y_pred[i] = a + b*x[i]
  return y_pred
```

```python
Y_pred = pred(a,b,n,X)
print(Y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_pred,Y)
```

```python
plt.plot(Y_pred,Y,color='red')
plt.title("Plotting Y-Predicted vs Y-Actual")
plt.xlabel("Y-Predicted")
plt.ylabel("Y-Actual")
plt.show()
```
